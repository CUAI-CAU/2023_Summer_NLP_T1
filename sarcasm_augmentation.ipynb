```python
# Import libraries 
import os
import openai
from transformers import pipeline
import time

classifier = pipeline(
    "sentiment-analysis", model="sangrimlee/bert-base-multilingual-cased-nsmc",max_length=512,truncation=True
)


path = 'C:/Users/82109/Downloads/cuai/TS_01. KAKAO(1)/'
save = 'C:/Users/82109/Downloads/cuai/save/'
files = os.listdir(path) 
```

```python
def paraphrase():
    
    content = ''
    prompt = f'{text} Paraphrase above dialog still in Korean. And be sure to keep output form with original dialog form'
    
    messages = [
                {'role': 'system', 'content': 'You are a helpful assistant.'},
                {'role': 'user', 'content': prompt},
            ]

    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo',
        messages=messages ,
        temperature = 0.5
    )

    paraphrased = (response['choices'][0]['message']['content'])

    time.sleep(1)
    
    return paraphrased    
```

```python
import re

def turn_dialog(t) :
    sentences = [p for p in re.findall("'([^']*)'",paraphrased) if p!=""]
    i = -1
    
    text = []
    
    i = len(sentences)-1
    for turn in range(0,t) :
        if i==0 :
            text.append(sentences[i])
            break
        while sentences[i][0]==sentences[i-1][0] :
            text.append(sentences[i])
            i-=1
        text.append(sentences[i])
    return text[::-1]
```

```python
def sarcasm_generation():

    prompt = f"{paraphrased_turn} 제시한 대화문의 전체 맥락을 파악하고 마지막 문장에 대해 비꼬는 반어법(sarcastic) 문장을 생성해줘. 이 때 출력 결과는 반드시 한 문장으로 생성해줘. 그리고 입력 형태와 동일한 형태로 발화자를 명시해줘. 예를 들어 생성한 대답의 발화자가 2라면 '2 : 생성 문장'과 같이 생성해줘. 그리고 주어진 대화의 마지막 발화자랑 생성한 대답의 발화자는 다르게 생성하고 이전 문장을 반복해서는 안돼. 반어법의 예시를 줄게. {{ 1: \"계란프라이가 바싹 타버렸어\", 2:\"이거 정말 바삭바삭하겠는걸\" }}의 문장에서 발화자 2와 같이 실제로 계란프라이가 바싹 타서 먹을 수 없음에도 불구하고 바삭바삭하다고 칭찬하는 것처럼 비꼬는 형태야 " 
    
    messages = [
                    {'role': 'system', 'content': 'You are a funny man has a great sense of humor.'},
                    {'role': 'user', 'content': prompt }
    ]
        
    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo',
        messages=messages
    )

    sarcasm = (str(response['choices'][0]['message']['content']).strip())
    
    return sarcasm
```
